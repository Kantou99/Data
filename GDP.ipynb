```python
import matplotlib.pyplot as plt

# Data for reproduction
countries = [
    "Nigeria", "Cte dIvoire", "Ghana", "Cameroon", "Senegal", "Guinea", "Mali",
    "Gabon", "Burkina Faso", "Benin", "Congo (Brazzaville)", "Niger", "Chad",
    "Equatorial Guinea", "Mauritania", "Togo", "Liberia", "Sierra Leone",
    "Cape Verde", "Central African Republic", "The Gambia", "So Tom & Prncipe"
]
gdp_values = [
    364.0, 78.0, 75.0, 48.0, 33.0, 22.0, 21.5, 21.0, 20.0, 20.0, 15.0, 15.0, 13.0,
    12.0, 10.5, 9.0, 5.0, 4.0, 2.0, 2.0, 2.0, 1.0
]

# Reverse data for horizontal bar chart
countries.reverse()
gdp_values.reverse()

# Create the horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(countries, gdp_values, color='skyblue', edgecolor='black')
plt.xlabel('Gross Domestic Product (US$ bn; market exchange rates)', fontsize=12)
plt.title('Gross Domestic Product by Country', fontsize=14)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# prompt: Could put dark blue on nigeria and ash colour of ohers

import matplotlib.pyplot as plt

# Data for reproduction
countries = [
    "Nigeria", "Cte dIvoire", "Ghana", "Cameroon", "Senegal", "Guinea", "Mali",
    "Gabon", "Burkina Faso", "Benin", "Congo (Brazzaville)", "Niger", "Chad",
    "Equatorial Guinea", "Mauritania", "Togo", "Liberia", "Sierra Leone",
    "Cape Verde", "Central African Republic", "The Gambia", "So Tom & Prncipe"
]
gdp_values = [
    364.0, 78.0, 75.0, 48.0, 33.0, 22.0, 21.5, 21.0, 20.0, 20.0, 15.0, 15.0, 13.0,
    12.0, 10.5, 9.0, 5.0, 4.0, 2.0, 2.0, 2.0, 1.0
]

# Reverse data for horizontal bar chart
countries.reverse()
gdp_values.reverse()

# Create a color list with dark blue for Nigeria and ash for others
colors = ['darkblue' if country == 'Nigeria' else 'lightgray' for country in countries]

# Create the horizontal bar chart
plt.figure(figsize=(5, 8))
plt.barh(countries, gdp_values, color=colors, edgecolor='black')
plt.xlabel('Gross Domestic Product (US$ bn; market exchange rates)', fontsize=12)
plt.title('Gross Domestic Product by Country', fontsize=14)
plt.tight_layout()

# Display the chart
plt.show()
```
